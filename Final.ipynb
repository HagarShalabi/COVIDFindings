#### Hypothesis Test-T Test

```python
#) Whats the relationship between the Country's Vaccination rate and its CO2 emissions? 
```

```python
##Get the 2 mean values for the 2 attributes
x1Mean=Data['CO2 Emissions (tonnes)'].mean()
x2Mean=Data['VaccinationRate'].mean()
x1Mean,x2Mean
```

```python
##Get the 2 variance values for the 2 attributes
x1Var=Data['CO2 Emissions (tonnes)'].var(ddof=1)
x2Var=Data['VaccinationRate'].var(ddof=1)
x1Var,x2Var
```

```python
#Formulating the equation
A1=len(Data['CO2 Emissions (tonnes)'])
A2=len(Data['VaccinationRate'])
ANS=(A1-1)*x1Var+(A2-1)*x2Var
DOF=A1+A2-2
part1=ANS/DOF
part2=(1/A1)+(1/A2)
SCORE=np.sqrt(part1*part2)
SCORE
```

```python
#Getting tscore
tScore=(x1Mean-x2Mean)/SCORE
tScore
```

```python
#Comparing tscore to the table value
import scipy
Value=scipy.stats.t.ppf(1-0.05, DOF)
tScore, Value
```

#### The t score>tabular value, the null hypothesis is rejected. The attributes are dependent on one another. Both attributes have a relationship.  <-- T score is too big, thats suspicious.
